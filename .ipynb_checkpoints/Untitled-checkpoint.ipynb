# Introduction

Il s'agit d'un gestionnaire de tâche. Une tâche dispose d'un identifiant, d'un titre, d'une description, d'une date de début et de fin, d'un status, d'une priorité, de commentaires et de sous-tâches nécessaires à son exécution.
J'expliquerai d'abord l'utilisation du programme, puis le fonctionnement du code en lui même dans une documentation.

# Utilisation

L'usage du programme est très simple et guidé. Tout se fait depuis le terminal, en question-réponse étape par étape. On compile et exécute avec la commande :
`++ -c main.cpp Priority.cpp Status.cpp Task.cpp Date.cpp fichier_texte.cpp User_interface.cpp && g++ -o ack main.o Priority.o Status.o Task.o Date.o fichier_texte.o User_interface.o && ./ack`

Le programme écrit dans le terminal :
`Quelle action désirez vous accomplir : create, change, delete, show ?`

`create` permet de créer une nouvelle tâche, `change` d'en modifier une, `delete` d'en supprimer une et `show` d'en afficher.

### 1 : create

On demande alors successivement, en prenant une réponse à la fois : `Titre (sans point virgule) :` ; `Description (sans point virgule) :` ; `Priorité (no problem, think about it, kinda pressing but ok, get to work now, urgent):` ; `Status (open, closed, in progress, forgotten long ago, friendly):` ; `Sous-tâches (identifiants séparés par des espaces):` ; `Combien de commentaires ?` ; puis `Commentaire : ` autant de fois qu'on vient de le demander.

Il est possible de ne rien répondre (ne pas mettre de commentaire par exemple, ou de status).

Après toutes ces questions, on renvoie un message de confirmation : `La tâche a bien été créée.` et le programme termine.

Exemple :

`g++ -c main.cpp Priority.cpp Status.cpp Task.cpp Date.cpp fichier_texte.cpp User_interface.cpp && g++ -o ack main.o Priority.o Status.o Task.o Date.o fichier_texte.o User_interface.o && ./ack`

Sinon : `Element à modifier (date de fin, priorité, status, progrès, sous-tâches, commentaires):`

# Fonctionnement du code

## Fonctionnement général du code

J'ai décidé de stocker les tâches dans un fichier texte, nommé `task_repository.txt`, à raison d'une tâche par ligne. Pour les créer/modifier/etc... j'ai créé une classe Task, et pour ses attributs, les classes Priority, Status et Date.

Mon code comporte 8 fichiers. `functions.hpp` contient le squellette de chaque classe et fonction, `Task.cpp` , `Status.cpp` , `Priotity.cpp` , et `Date.cpp`  contiennent les codes de chacune des classes correspondantes. `User_interface.cpp` contient les fonctions de questions et d'affichage pour interagir avec l'utilisateur, et `fichier_texte.cpp` contient toutes les fonctions qui écrivent ou lisent dans le fichier texte. Le dernier est `main.cpp`.

Le fonctionnement est le suivant :

Pour créer une tâche, on demande tous les attributs, on crée l'objet Task correspondant et on l'écrit.

Pour changer/supprimer une tâche, on crée un nouveau fichier texte `new_repository.txt`. On lit `task_repository.txt` ligne 
à ligne. A chaque ligne, on crée l'objet Task correspondant, et on teste si c'est celui qu'on cherche. Si non, on recopie la ligne dans `new_repository.txt`. Si oui, pour supprimer, on ne recopie pas la ligne et on enchaîne. Pour changer, on modifie l'objet Task, et écrit la ligne correspondante et on enchaîne. A la fin, on supprime `task_repository.txt` et on donne son nom à `new_repository.txt`.

Pour afficher, on parcourt `task_repository.txt` en créant les tâches ligne à ligne, et on affiche au passage celle qui nous intéressent.

## Fonctionnement détaillé 

Dans `task_repository.txt` on écrit les tâches à raison d'une par ligne sous la forme : 0_id ; 1_title ; 2_descr ; 3_date_creation ; 4_date_end ; 5_priority ; 6_status ; 7_progress ; 8_sous-tâches ; 9_commentaires

Exemple du contenu de `task_repository.txt` :

`11;lire;finir le cycle de fondation;2/2/2020;19/3/2020;no problem;closed;100;12;|lire les tomes dans l'ordre|les acheter en anglais
14;sortir le chien;aller au parc;27/3/2020;27/3/2020;think about it;closed;100;;|prendre la laisse
22;faire les courses;eau, pain, fromage;31/3/2020;31/3/2020;no problem;open;0;14 15;|eau évian|pain frais`


# Documentation

Je suivrai ici l'odre des fonctions dans `functions.cpp`.

## Date.cpp

### Class Date

Une Date a 3 attributs qui sont des entiers et correspondent à jour, mois, année.

##### Constructeurs :
 
`Date()` crée la date du jour-même. Ce constructeur est utilisé pour initialiser les dates de débuts et de fin lors de la création de tâches. Il fait appelle à des fonctions de la librairie `<ctime>`. Utilisé dans le constructeur `Task`.

`Date (int d, int m, int y)` crée la date d/m/y, qui sert pour modifier une date de fin. Utilisé dans `Date demanderDate ()` et dans `Task text_to_task (string text)`.


##### Méthodes :

`void print_date()` affiche la date dans le terminal au format day/month/year.

`string write_date()` renvoie la date sous forme de string "day/month/year". Utilisée dans `string Task::write ()`.

### Fonction

`Date text_to_date (string date)` prend une date sous forme de string "day/month/year" et crée l'objet Date correspondant.

## Status.cpp

### Class Status

Un `Status` a un attribut `sta` de type `string`, qui ne peut être que parmi la liste prédéterminée dans le tableau `tab_stati`. Il y avait initialement une fonction supplémentaire qui renvoyait un message d'erreur si on essayait de créer un status avec autre chose que ceux autorisés, mais il envoyait beaucoup de messages parasites quand on travaillait avec des tâches sans status.

##### Constructeurs :

`Status ()` est juste un constructeur vide qui crée un status vide. Utilisé dans le constructeur `Task`.

`Status (string status_in)` vérifie si `status_in` est un vrai status et le crée si c'est le cas. Sinon, il crée juste un status vide. Utilisé dans `Task text_to_task (string text)`.

##### Méthodes

`void print_status ()`  affiche le status dans le terminal.

`string write_status ()` renvoie le status. Utilisée dans `string Task::write ()`.

## Priority.cpp

### Class Priority

Une `Priority` a un attribut `prio` de type `string`, qui ne peut être que parmi la liste prédéterminée dans le tableau `tab_priorities`. Il y avait initialement une fonction supplémentaire qui renvoyait un message d'erreur si on essayait de créer une priorité avec autre chose que celles autorisées, mais il envoyait beaucoup de messages parasites quand on travaillait avec des tâches sans priorité.

##### Constructeurs :

`Priority ()` est juste un constructeur vide qui crée une priorité vide. Utilisé dans le constructeur `Task`.

`Priority (string priority_in)` vérifie si `priority_in` est une vraie priorité et la crée si c'est le cas. Sinon, il crée juste une priorité vide. Utilisé dans `Task text_to_task (string text)`.

##### Méthodes :

`void print_priority ()`  affiche la priorité dans le terminal.

`string write_priorité ()` renvoie la priorité. Utilisée dans `string Task::write ()`.

## Task.cpp

Une tâche a 10 attributs. L'identifiant, le titre et la description sont en privé car on ne va pas les modifier. Les autres sont publics. Avec plus de temps j'aurais créé des classes pour les sous-tâches et les commentaires pour rendre le code plus lisible, mais ces attributs ont été ajoutés après tout le reste et n'ont pas eu droit à leurs classes, ce qui expliquent certaines lourdeurs dans les fonctions les concernant.

##### Constructeurs :

`Task (int id_in, string title_in, string descr_in, Date date_debut = Date(), Date date_fin = Date(),Priority prio = Priority(), Status sta = Status(), int prog = 0, string sub_ta = "", string com = "")` crée la tâche à partir d'attributs potentiellement non renseignés. Seuls l'identifiant, le titre et la description sont nécessaires. Il utilise les constructeurs vides `Date()` `Priority()` et `Status()`.

##### Méthodes :

`int get_task_id ()`, `string get_title ()` et `string get_descr ()` renvoient respectivement l'identifiant, le titre et la description. `int get_task_id ()` est utilisée dans `void show_id (int id)`.

`void print_task ()` affiche la tâche dans le terminal. Utilisée dans `void show_id (int id)`, `void show_priority (Priority prio)` et `void show_status (Status sta)`.

`string write ()` écrit la tâche sous forme d'une ligne de texte, qui est appelée à être écrite dans le fichier texte. Utilise `write_date()`, `write_priority()` et `write_status()`. Il y a une forme de redondance avec `write_task`, cela vient de leur utilisation dans le `main`, selon que le fichier d'écriture est déjà ouvert ou non. `write_task` est utilisée uniquement lorsqu'on crée une nouvelle tâche, `string Task::write ()` uniquement pour les modifications.

## User_interface.cpp

Ici sont les fonctions qui posent les questions et affichent les réponses à l'utilisateur.

`string demanderGeneral (string question)` pose la question `question` et renvoie la réponse de l'utilisateur sous forme de `string`. `int demanderId ()` et `Date demanderDate ()` sont deux cas particuliers car ils convertissent directement la string de réponse en le type attendu. `Date demanderDate ()` utilise le constructeur `Date (int d, int m, int y)`.

`void print_com (string com)` affiche les commentaires d'une tâche prédéterminée. `com` est la partie commentaire d'une ligne du fichier texte.

`void show_id (int id)` affiche la tâche d'identifiant `id`. Utilise `Task text_to_task (string text)`, `int get_task_id ()` et `void print_task ()`.

`void show_priority (Priority prio)` affiche toutes les tâches de priorité `prio`. Utilise `Task text_to_task (string text)` et `void print_task ()`.

`void show_status (Status sta)` affiche toutes les tâches de status `sta`. Utilise `Task text_to_task (string text)` et `void print_task ()`.


## fichier_texte.cpp

Contient toutes les fonctions qui ont pour vocation d'écrire ou de modifier le fichier texte `task_repository.txt`.
La méthode est toujours la même et a été expliquée dans la partie "Fonctionnement général du code".

`void write_task (Task task)` écrit une tâche dans le fichier texte, selon le format expliqué dans la partie "Fonctionnement détaillé". Utilise `int get_task_id ()`, `string get_title ()` et `string get_descr ()`, `write_date()`, `write_priority()` et `write_status()`. Il y a une forme de redondance avec `string Task::write ()`, cela vient de leur utilisation dans le `main`, selon que le fichier d'écriture est déjà ouvert ou non. `write_task` est utilisée uniquement lorsqu'on crée une nouvelle tâche, 
`string Task::write ()` uniquement pour les modifications.

`int get_id ()` donne l'identifiant de la prochaine tâche, c'est-à-dire le nombre de tâches déjà crées (et potentiellement supprimée) +1. Pour pouvoir garder le compte, il faut avoir l'information localisée en dehors du programme. Je la stocke dans le fichier `identifiant.txt` qui ne contient qu'un entier. A chaque création de tâche, on récupère cet entier, on crée le fichier `new_identifiant.txt` dans lequel on écrit le nouvel entier, on supprime `identifiant.txt` et on donne son nom à `new_identifiant.txt`.

`Task text_to_task (string text)` prend une ligne du fichier texte (qui correspond donc à une tâche) et crée l'objet Task correspondant. Utilise `Date text_to_date (string date)` et les constructeurs `Status (string status_in)` et `Priority (string priority_in)` pour créer la tâche.

Toutes les fonctions suivantes changent l'attribut de leur nom dans la tâche d'identifiant `id` : 
`void change_end_date (int id)` 
`void change_priority (int id)`
`void change_status (int id)`
`void change_progress (int id)`
`void change_sub_task (int id)`
`void change_com (int id)`
Le fonctionnement est toujours le même, c'est celui décrit dans la partie "Fonctionnement général du code", en passant par un nouveau fichier texte dans lequel on recopie toutes les tâches sauf celle à modifier que l'on modifie au préalable en passant par les méthodes de la classe Task.

`void delete_task (int id)` supprime la tâche d'indice `id` du fichier texte.

`bool exist_task (int id)` renvoie `true` si la tâche d'indice `id` existe et `false` sinon.

`bool est_dedans (int x, int* t, int len)` prend un tableau d'entier `t` de longueur `len` et renvoie `true` si `x` est dans `t` et `false` sinon. J'ai cherché des fonctions C++ qui faisaient cela, j'en ai trouvé qui marchaient presque, mais je n'ai pas réussi à trouver quelque chose de suffisamment simple par rapport à juste recoder la fonction dans ce cadre particulier.
